```python
import pandas as pd
```

```python
rent_data = pd.read_csv("data/ourbudget - rents.csv")
```

```python
home_data = pd.read_csv("data/ourbudget - homeitems.csv")
```

```python
amounts_dict = {
    "Atharva": {},
    "Hritik": {},
    "Sanket": {},
    "Darshan": {}   
}
```

```python
for item in amounts_dict.keys():
    amounts_dict[item]["transactions"] = []
```

```python
amounts_dict
```

```python
for index, row in home_data.iterrows():
    amount = row["how much was spent"]
    if row["consumed by everyone"] == "y":
        consumers = list(amounts_dict.keys())
        consumers.remove(row["who spent it"])
        
        for consumer in consumers:
            amounts_dict[consumer]["transactions"].append({
                "amount": row["how much was spent"]/4,
                "purpose": row["where was the money spent"],
                "paid-by": row["who spent it"],
                "Date": row["Date"]
            })
            
    if row["consumed by everyone"] == "n":
        for consumer in ['Atharva', 'Hritik', 'Sanket', 'Darshan']:
            if not pd.isnull(row[consumer + " Share"]) and row[consumer + " Share"] != 0:
                if row["who spent it"] != consumer:
                    amounts_dict[consumer]["transactions"].append({
                    "amount": row[consumer + " Share"],
                    "purpose": row["where was the money spent"],
                    "paid-by": row["who spent it"],
                    "Date": row["Date"]
                })
        
```

```python
user_df = {}
```

```python
for consumer in ['Atharva', 'Hritik', 'Sanket', 'Darshan']:
    data = amounts_dict[consumer]["transactions"]
    tmp_df = pd.DataFrame.from_dict({
        "Date": [item["Date"] for item in data],
        "Purpose": [item["purpose"] for item in data],
        "Amount": [item["amount"] for item in data],
        "Paid by": [item["paid-by"] for item in data]
    })
    user_df[consumer] = tmp_df
```
